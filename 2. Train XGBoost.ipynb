Grid Search Method

```python
param_grid = {'learning_rate':[0.003, 0.01,0.03],
             'n_estimators':[400, 800, 1200],
             'subsample':[0.5, 0.6, 0.7],
             'max_depth':[3, 4, 6],
             'reg_lambda':[1, 3,4, 5]}
fit_params = {#'early_stopping_rounds':10,
             'eval_metric':'logloss'}

model_grid = xgb.XGBClassifier(use_label_encoder=False)
kfold = sk.model_selection.StratifiedKFold(n_splits=3, shuffle=True)
grid_search = sk.model_selection.GridSearchCV(model_grid, param_grid, scoring='roc_auc', n_jobs=-1, cv=kfold, verbose=2, return_train_score=True)
grid_result = grid_search.fit(X_train, Y_train, **fit_params)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
train_means = grid_result.cv_results_['mean_train_score']
sorted_means=means.copy()
sorted_means.sort()
top_20 = sorted_means[-20]
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, train_mean, param in zip(means, stds, train_means, params):
    if mean >= top_20:
        print("Val: %f (%f), Train: %f with: %r" % (mean, stdev, train_mean, param))
```
